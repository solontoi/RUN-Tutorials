## 1. Orbital Elements

Orbital elements are like  the "address" of a celestial object in space. For this tutorial we will be using orbital elements that describe the **size**, the **shape** and the **tilt** of an asteroid's elliptical orbit around the Sun.

| Symbol | Name | What it is | Tells you |
| --- | --- | --- | --- |
|a|Semi-major axis| Half the longest diameter of the ellipse, it is also the average distance between the object and the Sun.|The size of the orbit.|
|e|Eccentricity|The ratio of the distance between the foci to the major axis, a measure of how much the orbit deviates from a perfect circle (where $e = 0$).|The shape (roundness) of the orbit.|
|i|Inclination|The angle between the orbital plane and the Ecliptic (the plane of Earth's orbit about the Sun).|The tilt of the orbit.|
|q|Perihelion|The point in the orbit where the object is closest to the Sun.|The minimum distance from the Sun, $q=a(1−e)$|
|Q|Aphelion|The point in the orbit where the object is farthest from the Sun.|The maximum distance from the Sun $Q=a(1+e)$|

##### ***a*** - the *semi-major axis*.
  Geometrically this is the semi-major axis (one half the long axis) of the elipse that the object's orbit traces, and is also the *average* distance of the object from the Sun over the course of its orbit. Here it is measured in *astronomical units*, (au),  with $1$ au being approximately the average Earth-Sun distance, and specifically $1$ au = $149,597,870,700$ m.

##### ***e*** - the *eccentricity*.
  Geometrically this is the *eccentricity* of the ellipse, a measure of how much an conic section (generally an ellipse) deviates from being circular:
- The eccentricity of a circle is $0$.
- The eccentricity of a non-circular ellipse is between $0$ and $1$.
- The eccentricity of a parabola is $1$.
- The eccentricity of a hyperbola is greater than $1$.

##### *q* - the *perihelion* and *Q* - the *aphelion*
The perihelion (*q*) and aphelion (*Q*) are the nearest and farthest points, respectively, of a body's orbit around the Sun, and may be computed from *a* and *e* as follows:
- Perihelion: $q = a~ (1 - e)$
- Aphelion: $Q = a~ (1 + e)$

##### ***i*** - the *inclination*.
  This is angle between the object's orbital plane and the the ecliptic, the plane in which the Earth orbits the Sun. It is measured in degrees with $0^{\circ}$ being coplanar with the Earth's orbit, in the same direction, or "prograde", and by the time the inclination hits $180^{\circ}$ the orbit is again in the plane of the Earth's, but orbiting in the reverse direction, or "retrograde."

Due to the often used letter `i` in python, we'll be using an abbreviation like `incl` for *inclination* in our code

## 2. MPC-based Dynamical Orbital Groups

following the orbit type definitions from the <a href="https://www.minorplanetcenter.net/mpcops/documentation/orbit-types/">Minor Planet Center</a> we classify orbits into the following types:

- Near Earth Objects:
    - `AMO` = Amor NEO
    - `APO` = Apollo NEO
    - `ATE` = Aten NEO
    - `IEO` = inner-Earth object NEO
- `MCA` = Mars-crosser asteroid
- Main Belt Asteroids:
    - `IMB` = inner main-belt asteroid
    - `MBA` = main-belt asteroid
    - `OMB` = outer main-belt asteroid
- `TJN` = Jupiter Trojan, 
- `CEN` = Centaur
- `TNO` = transneptunian object
- `PAA` = parabolic asteroid
- `HYA` = hyperbolic asteroid

#### 2.1.1 Amors

Amors, named after asteroid 1221 Amor, come within 1.3 au of the Sun, making them NEOs, but stay *outside* of the orbit of the Earth.

An asteroid is in the Amor group if:
- The orbital semi-major axis, $a$, is greater than $1.0$ au ($a > 1.0$ au)
- The orbit does not cross that of Earth meaning the asteroid' perihelion is greater than Earth's aphelion, so $q > 1.017$ au
- The object is a near-Earth object (NEO) and such $q < 1.3$ au.

#### 2.1.2 Apollos

Apollos, named after asteroid 1862 Apollo, come within 1.3 au of the Sun, making them NEOs, and *cross* the orbit of the Earth, with semi-major axes *greater* than Earth's

An asteroid is in the Apollo group if:
- The orbital semi-major axis, $a$, is greater than $1.0$ au ($a > 1.0$ au)
- The orbit crosses that of Earth, so the asteroid's perihelion is less than Earth's aphelion, so $q < 1.017$ au
- The object is a near-Earth object (NEO) and such $q < 1.3$ au.

#### 2.1.3 Atens

Atens, named after asteroid 2062 Aten, come within 1.3 au of the Sun, making them NEOs, and *cross* the orbit of the Earth, with semi-major axes smaller than Earth's

An asteroid is in the Aten group if:
- The orbital semi-major axis, $a$, is less than $1.0$ au ($a < 1.0$ au)
- The orbit crosses that of Earth so the asteroid's aphelion is less than Earth's perihelion, so $Q > 0.983$ au
- The object is a near-Earth object (NEO) and such $q < 1.3$ au.

#### 2.1.4 inner-Earth objects (IEO or Atiras)

Inner-Earth Objects often called "Atiras" after the asteroid 163693 Atira, and NEO's whose orbit is entirely inside that of the Earth's, with aphelion less than the Earth's perihelion.

An asteroid is in the inner-Earth group if:
- The orbital semi-major axis, $a$ is less than $1.0$ au ($a < 1.0$ au)
- The orbital aphelion distance is less than the Earth's perihelion, so the obit is always entirely inside the orbit of the Earth ($Q < 0.983$).

### 2.2 Mars-crosser asteroids (MCA)

Mars-crosser asteroids are asteroids that cross the orbit of Mars. Following the MPC, we will consider asteroids with a semi-major axis of less than $3.2$ au, and with a minimum distance to the Sun of more than $1.3$ au that cross the orbit of Mars to be in this group.

An asteroid is in the Mars-crosser asteroid group is:
- The orbital semi-major axis, $a$ is less than $3.2$ au ($a < 3.2$ au)

- The orbital perihelion distance is between the NEO limit of $1.3$ au and the aphelion of Mars, $1.666$ au. ($1.3$ au $< q < 1.666$ au)

### 2.3. Main Belt Asteroids

The following diagram from is a visualization of orbital group "zones" in the middle Solar System in the realm between mars (red orbit) and Jupiter (yellow orbit)

<center><img src="./inner_solar_system_orbits.jpg" width="600"/></center>

The main asteroid belt is a large collection of asteroids spanning a zone from around $2$ au to $3.2$ au, the asteroids in this zone we will consider to be *main-belt asteroids* for this tutorials. The asteroids beyond Mars, but not in this classical main belt region to be *inner main-belt asteroids*, and those beyond the classical belt but still sufficiently inside of Jupiter's orbit we'll consider *outer main-belt asteroids*.

Note that quite often the main-belt itself is divided into an inner, middle, and outer section, but this is not what is being referenced in this tutorial.

#### 2.3.1 inner main-belt asteroid (IMB)

These are the asteroids beyond the realm of Mars-crossers, but inside the classical main belt. An asteroid is part of the inner main-belt group if:
- The orbital semi-major axis, $a$ is less than $2.0$ au, iniside of the classical asteroid belt ($a > 2.0$ au)

- The asteroid does not cross the orbit of mars so the asteroid's orbital perihelion distance is greater than the aphelion of Mars, $1.666$ au (q > 1.666$ au)

#### 2.3.2 main-belt asteroid (MBA)

These are asteroids that are in the classically-defined main asteroid belt with semi-major axes between $2.0$ and $3.2$ au, orbiting entirely outside the orbit of Mars. An asteroid is a part of the main-belt asteroid group if
- The orbital semi-major axis, $a$ is within of the classical asteroid belt ($2.0$ au $< a < 3.2$ au)
- the closest it nears the Sun is outside the farthest distance that Mars does. ($q > 1.666$ au)

#### 2.3.3 outer main-belt asteroid (OMB)

These are the asteroids just outside the classical asteroid belt (beyond the 2:1 resonance with Jupiter), but sufficiently inside the orbit of Jupiter. An asteroid is part of the outer main-belt group if:
- The orbital semi-major axis, $a$ is greater than $3.2$ au, outside of the classical asteroid belt, but still sufficiently inside the orbit of Jupiter ($3.2 < a < 4.6$)

### 2.4. Jupiter Trojans (TJN)

Jupiter Trojans (TJN) are objects that shares Jupiter's orbit around the sun. Most are located in the leading and trailing Lagrange points of Jupiter's orbit about the sun, $60^{\circ}$ ahead and behind Jupiter in its orbit. An asteroid is part of the Jupiter Trojans group if:
- The orbital semi-major axis, $a$ is sufficiently close to Jupiters ($4.6$ au $< a < 5.5$ au)
- The orbital eccentricity is less than $0.3$ ($e < 0.3$)

### 2.5 Centaurs (CEN)

a centaur is a small Solar System body that orbits the Sun between Jupiter and Neptune, so an object is a part of the Centaur group if
- The semi-major axis, $a$ is sufficiently outside that of Jupiter ($a > 5.5$)
- The semi-major axis, $a$, is inside the orbit of Neptune ($a < 30.1$)

### 2.6 Transneptunian objects (TNO)


Transneptunian objects are any objects in the Solar System that orbit the Sun at a greater average distance than Neptune, so an object is part of the Transneptunian group if

-The semi major axis, $a$ is larger than that of Neptune ($a > 30.1$)

### 2.7 Parabolic asteroids (PAA)

Parabolic asteroids have orbits that are parabolas, and hence are barely unbound. To be part of the parabolic asteroid group:

- The orbital eccentricity, $e$, should be that of a parabola ($e = 1$)

### 2.8 Hyperbolic asteroids (HYA)

Hyperbolic asteroids have orbits that are hyperbolas, and hence are unbound. To be part of the hyperbolic asteroid group:

- The orbital eccentricity, $e$, should be that of a hyperbola ($e > 1$)
